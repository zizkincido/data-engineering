# Q3

```python
filtered_trips = trip_data[(trip_data['lpep_pickup_datetime'].dt.month == 11) & (trip_data['trip_distance'] <= 1)]
print(filtered_trips)
```

# Q4

```python
filtered_trips = trip_data[trip_data['trip_distance'] < 100]
max_trip_distance = filtered_trips['trip_distance'].max()
print(filtered_trips[filtered_trips['trip_distance'] == max_trip_distance]['lpep_pickup_datetime'])
```

# Q5

```python
filtered_trips = trip_data[(trip_data['lpep_pickup_datetime'].dt.month == 11) & (trip_data['lpep_pickup_datetime'].dt.day == 18)]

zone_totals = filtered_trips.groupby('PULocationID')['total_amount'].sum()

pickup_location_id = zone_totals.idxmax()
max_total_amount = zone_totals.max()

result = zone_data[zone_data['LocationID'] == pickup_location_id]
print(f"Pickup Zone: {result['Zone'].values[0]}")
print(f"Total Amount: ${max_total_amount:.2f}")
print(result)
```

# Q6

```python
east_harlem_north_id = zone_data[zone_data['Zone'] == 'East Harlem North']['LocationID'].values[0]

filtered_trips = trip_data[(trip_data['lpep_pickup_datetime'].dt.month == 11) & (trip_data['PULocationID'] == east_harlem_north_id)]

max_tip = filtered_trips[filtered_trips['tip_amount'] == filtered_trips['tip_amount'].max()]

result = zone_data[zone_data['LocationID'] == max_tip['DOLocationID'].values[0]]
print(f"Drop-off Zone: {result['Zone'].values[0]}")
```
